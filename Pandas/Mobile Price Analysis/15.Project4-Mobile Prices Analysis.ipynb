```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Question 1. Which devices have RAM (MB) size Greater than 6000 ?

highest_ram_devices = df[df['RAM (MB)'] > 6000]
plt.figure(figsize=(14, 8))
sns.barplot(x='Name', y='RAM (MB)', data=highest_ram_devices, palette='Spectral')
plt.title("Devices with RAM Greater than 6000 MB")
plt.xlabel("Device Name")
plt.ylabel("RAM (MB)")
plt.xticks(rotation=90) 
plt.show()
```

```python
# Question 2. What is the Name of the device with the largest battery capacity (mAh)?

battery = df.sort_values(by='Battery capacity (mAh)', ascending=False)
top_20_devices = battery.head(20)
print(top_20_devices[['Name', 'Battery capacity (mAh)']])
plt.figure(figsize=(14, 8))
plt.bar(top_20_devices['Name'], top_20_devices['Battery capacity (mAh)'], color='Pink')
plt.title('Top 20 devices with best battery capacity ',fontsize=25)
plt.xlabel('Device names',fontsize=15)
plt.ylabel('Battery capacity (mAh)',fontsize=15)
plt.xticks(rotation=90) 
```

```python
# Question 3. How many devices doesnot have a touchscreen?

touchscreen=df.query("Touchscreen =='No'")
device = touchscreen['Name'].count()
print(f"Devices that does not have touchscreen: {device}")
```

```python
#Question 4. What is the price of the device with the largest screen size (inches)?

sorted = df.sort_values(by="Screen size (inches)").head(25)
plt.figure(figsize=(14, 8))
sns.barplot(x='Screen size (inches)', y='Price', data=sorted , color='skyblue')
plt.title("Price vs. Screen Size",fontsize=25)
plt.xlabel("Screen Size (inches)",fontsize=20)
plt.ylabel("Price",fontsize=20)
plt.xticks(rotation=45)
plt.show()
```

```python
# Question 5. Which brand has the most number of devices ?

brand_count = df['Brand'].value_counts().head(30)
most_common_brand = brand_count.idxmax()
print("Brand with the most devices:", most_common_brand)
plt.figure(figsize=(14, 8))
sns.barplot(x=brand_count.index, y=brand_count.values, palette='flare')
plt.title('Number of Devices per Brand',fontsize=25)
plt.xlabel('Brand',fontsize=15)
plt.ylabel('Number of Devices',fontsize=15)
plt.xticks(rotation=90)  
plt.tight_layout()
```

```python
# Question 6.How many devices support Bluetooth?

bluetooth_devices = df.query("Bluetooth == 'Yes'")
b_count = bluetooth_devices['Name'].count()
print(f"Number of devices that support Bluetooth: {b_count}")
```

```python
# Question 7. Which device has the highest resolution in terms of Resolution x and Resolution y?

df['Total Resolution'] = df['Resolution x'] * df['Resolution y']
sorted_by_resolution = df.sort_values(by='Total Resolution', ascending=False)
top_20_high_resolution_devices = sorted_by_resolution.head(20)
print(top_20_high_resolution_devices[['Name', 'Total Resolution']])
```

```python
# Question 8. Which operating system is most common among the devices?

os = df['Operating system'].value_counts()
plt.figure(figsize=(14, 8))
sns.barplot(x=os.index, y=os.values, palette='muted')
plt.title("Distribution of Operating Systems Among Devices",fontsize=25)
plt.xlabel("Operating System",fontsize=15)
plt.ylabel("Number of Devices",fontsize=15)
plt.xticks(rotation=45)
plt.show()
```

```python
# Question 9. What is the smallest screen size (inches) ?

smallest_screen_size = df['Screen size (inches)'].min()
print("Smallest screen size (inches):", smallest_screen_size)
```

```python
# Question 10. What is the average price of devices ?

average_price = df['Price'].mean()
print("Average price of devices:", average_price)
```

```python
# Question 11.Which device have price greater than 25000 and brand -samsung ?

a = df.query("Brand == 'Samsung' and Price > 25000")
plt.figure(figsize=(12, 6))
sns.barplot(x='Name', y='Price', data=a, palette='pastel')
plt.title('Samsung Devices with Price Greater Than 25000',fontsize=25)
plt.xlabel('Device Name',fontsize=15)
plt.ylabel('Price (in INR)',fontsize=15)
plt.xticks(rotation=90)
plt.tight_layout()
```

```python
# Question 12. Top most 10 expensive device and what brand is it?   

top_10_expensive = df.sort_values(by='Price', ascending=False).head(10)[['Name', 'Brand', 'Price']]
print("Top 10 Most Expensive Devices:\n", top_10_expensive)
plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_expensive['Name'], y=top_10_expensive['Price'], hue=top_10_expensive['Brand'], palette='flare')
plt.title('Top 10 Most Expensive Devices',fontsize=25)
plt.xlabel('Device Name',fontsize=15)
plt.ylabel('Price (INR)',fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

```python
# Question 13. what is the total number of models in the dataset ?

total_model = df['Model'].nunique()
print("Total number unique brands are:",total_model)
```

```python
# Question 14. Name of the devices that have Internal storage (GB) greater than or equal to 128 and brands.

dev = df.query("`Internal storage (GB)` >= 128")[['Name', 'Brand', 'Internal storage (GB)']].head(20)
print("Devices with internal storage >= 128 GB:")
print(dev)
plt.figure(figsize=(20, 10))
sns.barplot(x="Name", y="Internal storage (GB)",  hue="Brand",data=dev, palette="dark")
plt.title("Devices with Internal Storage >= 128 GB", fontsize=25)
plt.xlabel("Device Name", fontsize=15)
plt.ylabel("Internal storage (GB)", fontsize=15)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Question 15. How many devices have Operating system-iOS ?

os = df.query("`Operating system` == 'iOS'")
os_count = os['Name'].count()
print(f"Number of devices that have operating system iOS: {os_count}")
```

```python
# Question 16. Number of Devices that doesnot have GPS in the devices.

g = df.query("`GPS` == 'No'")
g_count = g['Name'].count()
print(f"Number of devices that have operating system iOS: {g_count}")
```

```python
# Question 17. total number of brands in the dataset ? 

total_brands = df['Brand'].nunique()
print("Total number unique brands are:",total_brands)
```

```python
# Question 18. Name the brand that have 4G/ LTE and 3G in the devices?

a = df.query("`4G/ LTE` == 'Yes' & `3G` == 'Yes'").head(15)
brand_counts = a ['Brand'].value_counts()
plt.figure(figsize=(12, 8))
sns.barplot(x=brand_counts.index, y=brand_counts.values, palette='viridis')
plt.title('Brands with 3G and 4G/LTE Support')
plt.xlabel('Brand')
plt.ylabel('Number of Devices')
plt.xticks(rotation=90)
plt.show()
```
