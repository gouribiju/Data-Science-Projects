```python
df['winner'].fillna('No Winner',inplace=True)
df['player_of_match'].fillna('No Player',inplace=True)
df['city'].fillna('Unknown', inplace=True)
```

```python
df.isnull().any()
```

```python
# Question 1 . Which team has got more number of toss_winner ?

import matplotlib.pyplot as plt
import numpy as np

toss=df['toss_winner'].value_counts()
plt.figure(figsize=(15, 6))
a=plt.bar(toss.index,toss.values,color='red',edgecolor='black')
plt.bar_label(a,padding=5, fontsize=20)
plt.title("Most Toss Winner Teams",fontsize=20)
plt.xlabel("Teams",fontsize=20)
plt.ylabel("Number of Toss Wins",fontsize=20)
plt.xticks(rotation=90)
```

```python
# Question 2. Which team has won most of the matches ?

a=df['winner'].value_counts()
plt.figure(figsize=(20,10))
b=plt.bar(a.index,a.values,color='skyblue')
plt.bar_label(b,padding=5,fontsize=15)
plt.title('Team with winning more times',fontsize=25)
plt.xlabel('Team',fontsize=20)
plt.ylabel('Number of times won',fontsize=20)
plt.xticks(rotation=90)
```

```python
# Question 3. Which teams won by_runs ?

a=df.query("win_by_runs > 100 ")
won_runs=a['winner'].value_counts()
plt.figure(figsize=(10, 6))
b=plt.bar(won_runs.index,won_runs.values,color='lightgreen')
plt.bar_label(b,padding=5, fontsize=10)
plt.title("Teams Won by runs",fontsize=25)
plt.xlabel("Teams",fontsize=15)
plt.ylabel("Number of times won by runs",fontsize=15)
plt.xticks(rotation=30)
```

```python
# Question 4. Count the number of matches that was assigned to 'AY Dandekar' as 1st umpire ?
 
a=df.query("umpire1=='AY Dandekar'")
a['umpire1'].value_counts()
```

```python
# Question 5. Which team has choosen least toss decision as field or bat ?

f=df.query("toss_decision=='field'")
Field=f['toss_decision'].value_counts()

b=df.query("toss_decision=='bat'")
Bat=b['toss_decision'].value_counts()

i=plt.bar(Field.index,Field.values,color='skyblue',edgecolor='black')
a=plt.bar(Bat.index,Bat.values,color='blue')

plt.bar_label(i,padding=2, fontsize=10)
plt.bar_label(a,padding= 2 , fontsize=10)
plt.title("Toss decision as field or bat choosen the team",fontsize=25)
plt.xlabel("Toss Decision",fontsize=15)
plt.ylabel("Number of batting  or fielding",fontsize=15)
```

```python
# Question 6. Count the number of matches that was held in the venue ' Rajiv Gandhi International Stadium, Uppal' ?
 
a=df.query("venue =='Rajiv Gandhi International Stadium, Uppal'")
a['venue'].value_counts()
```

```python
# Question 7. In which venue more number of matches were held ?

VENUE=df['venue'].value_counts().head(15)
plt.figure(figsize=(15,10))
a=plt.bar(VENUE.index,VENUE.values,color='yellow',edgecolor='black')
plt.bar_label(a ,padding=2, fontsize=15)
plt.title("More number of matches held in the venue",fontsize=20)
plt.xlabel("Teams",fontsize=20)
plt.ylabel("Venue",fontsize=20)
plt.xticks(rotation=90)
plt.tight_layout()
```

```python
# Question 8. Name the players that wons the most player of the match?

PLAYER=df['player_of_match'].value_counts().head(10)
plt.figure(figsize=(13, 6))
a=plt.bar(PLAYER.index,PLAYER.values,color='pink',edgecolor='black')
plt.bar_label(a,padding=5, fontsize=11)
plt.title("More number of times Man of the Match", fontsize=25)
plt.xlabel("Players", fontsize=15)
plt.ylabel("Number of Man of Match", fontsize=15)
plt.xticks(rotation=90)
```

```python
# Question 9. Which city played a least matches while comparing with all the season?

count_city = df['city'].value_counts().sort_values().head(15)
plt.figure(figsize=(15, 8))
a=plt.bar(count_city.index,count_city.values,color='maroon',edgecolor='black')
plt.bar_label(a,padding=5, fontsize=10)
plt.title("City which played least matches", fontsize=25)
plt.xlabel("city", fontsize=15)
plt.ylabel("Number of Match Played", fontsize=15)
plt.xticks(rotation=30)
```

```python
# Question 10. How many matches are ended as Tie?

a=df.query("result =='tie'")
a['result'].value_counts()
```

```python
# Question 11. Which team has made minimum toss decision ?

count = df['toss_winner'].value_counts().sort_values(ascending=True)
plt.figure(figsize=(18, 8))
a=plt.bar(count.index,count.values,color='orange',edgecolor='g')
plt.bar_label(a,padding=5, fontsize=10)
plt.title("Minimum toss decision ", fontsize=25)
plt.xlabel("Teams", fontsize=15)
plt.ylabel("Number of Toss Decision", fontsize=15)
plt.xticks(rotation=90)
```

```python
# Question 12. Which team combinations faced each other most frequently?

MATCHES = df.groupby(['team1', 'team2']).size().sort_values(ascending=False).head(10)
MATCHES
```

```python
# Question 13. Count the occurences of the matches that results in 'normal' ?

a=df.query("result =='normal'")
a['result'].value_counts()
```

```python
# Question 14. What is the total number of cities participated in the match?

total_city = df['city'].nunique()
print(f"Total number of countries in the dataset is: {total_city}")
```

```python
# Question 15.Which teams are won by wickets ?
    
a=df.query("win_by_wickets < 100 ")
won_wickets=a['winner'].value_counts()
plt.figure(figsize=(10, 6))
b=plt.bar(won_wickets.index,won_wickets.values,color='lightgreen')
plt.bar_label(b,padding=5, fontsize=10)
plt.title("Teams Won by wicket",fontsize=25)
plt.xlabel("Teams",fontsize=15)
plt.ylabel("Number of times won by wicket",fontsize=15)
plt.xticks(rotation=90)
```
